```python
#Точечные диаграммы используются, когда мы хотим показать связь между двумя переменными. 
#Диаграммы рассеяния иногда называют диаграммами корреляции, 
#потому что они показывают, как две переменные коррелируют.

df.plot(kind='scatter', 
        x='UnitPrice', 
        y='Country', 
        title='Correlation country/unitprice')
plt.xticks(rotation=0,ha='left')
plt.xlabel('UnitPrice',fontsize=20)
plt.ylabel('Country',fontsize=20)
plt.show()
```

```python
df.plot(kind='scatter', 
        x='Quantity', 
        y='Country', 
        title='Correlation country/quantity')
plt.xticks(ha='center')
plt.xlabel('Quantity',fontsize=20)
plt.ylabel('Country',fontsize=20)
plt.show()
```

```python
df.plot(kind='scatter', 
        x='Quantity', 
        y='UnitPrice', 
        title='Correlation quantity/unitPrice')
plt.xticks(ha='center')
plt.xlabel('Quantity',fontsize=20)
plt.ylabel('UnitPrice',fontsize=20)
plt.show()
#т.е чем ниже цена тем больше покупок, только две покупки по стоимости около 7000 усл.ед. (или это все же два выброса?!)
# данные с отрицательным количеством вариант 1 -отсечь и принимать во внимание
#вариант 2 - трансформировать в положительное, взять по модулю
```

```python
d_price=df['UnitPrice'].value_counts().to_frame().reset_index().rename(columns={'index':'UnitPrice'}) 
d_price.head(20)
#смотрим по какой стоимости были преобладающие покупки? это около 0,5 - 1 - 2 - 3 - 4 усл.ед. При том пик популярности товаров за 1-2 усл.ед.
#Соотвественно болью этого вида бизнеса онлайн интернетр магазина является низкая маржа в малостоимостных товарах.
#Видимо продают брелки/ручки/шарики - что-то совсем недорогое.
#Таким образом надо увеличивать средний чек, предлагать на страницах карточек товара сопуствующие товара "с этим покупают"
# Смотрим что в ценах за 7 усл.ед и 8 усл.ед разница на целую тысячу покупок, именно этот ценник считаем переломным.
```
