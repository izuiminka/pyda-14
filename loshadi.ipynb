```python
#Задание 3. Работа с пропусками
#Рассчитать количество пропусков для всех выбранных столбцов. 
#Принять и обосновать решение о методе работы с пропусками по каждому столбцу, 
#сформировать датафрейм, в котором пропуски будут отсутствовать.

'''Для численных(количественых) признаков можно воспользоваться методом принудительного заполнения пропусков. 
Например, на место пропуска можно записать среднее или медианное значение, полученное из остальных записей.
Считаю, что данные о температуре, пульсе и дыхании животных важны, возможно некоторые показатели для лошадей были в норме, 
поэтому при проставлении информации, они оказались пустыми; поэтому NaN-пропуски надо заполнить
 медианным значением столбца по их нормированным показателей. Также можно сгруппировать данные по значениям в столбцах, чтобы убедиться в структурированности данных '''
```

```python
df = pd.read_csv('horse_data.csv', na_values = '?') # еще раз читаем файл и делаем преобразования
df= df.rename(columns = {'38.50': 'temperature'})
df = df.rename(columns = {'66': 'puls'})
df = df.rename(columns = {'45.00': 'otherpuls'})
df = df.rename(columns = {'28': 'breath'})
df = df.rename(columns = {'8.40': 'eat'}) 
df.temperature.fillna(df.temperature.median(),inplace=True) # заменим Nan-пропуски в столбце температура медианой по температуре
df.puls.fillna(df.puls.median(),inplace=True) # заменим Nan-пропуски в столбце пульс медианой по пульсу
df.otherpuls.fillna(df.otherpuls.median(),inplace=True) # заменим Nan-пропуски в столбце другой пульс медианой по другому пульсу
df.breath.fillna(df.breath.median(),inplace=True) # заменим Nan-пропуски в столбце дыхание медианой по дыханию
```

```python
df.head(20)
```

```python
# рассчитаем вновь медианные значения по столбцам и скреднеквадратичное отклонение
print(df.temperature.median())
print(df.puls.median())
print(df.otherpuls.median())
print(df.breath.median())
print(df.temperature.std())
print(df.puls.std())
print(df.otherpuls.std())
print(df.breath.std())
```

```python
# Видим, что отклонения уменьшились,считаю, что данные приведены к более читабельному, чистому виду
```

```python
# ПРИМЕЧАНИЕ:ЕСЛИ ЗАМЕНИТЬ NaN НА ЗНАЧЕНИЕ МОДЫ, то почему значения такие же как при замене медианой?
df.temperature.fillna(df.temperature.mode()[0],inplace=True) # заменим Nan-пропуски в столбце температура медианой по температуре
df.puls.fillna(df.puls.mode()[0],inplace=True) # заменим Nan-пропуски в столбце пульс медианой по пульсу
df.otherpuls.fillna(df.otherpuls.mode()[0],inplace=True) # заменим Nan-пропуски в столбце другой пульс медианой по другому пульсу
df.breath.fillna(df.breath.mode()[0],inplace=True) # заменим Nan-пропуски в столбце дыхание медианой по дыханию
```

```python
# рассчитаем вновь значение моды по столбцам и скреднеквадратичное отклонение
print(df.temperature.mode()[0])
print(df.puls.mode()[0])
print(df.otherpuls.mode()[0])
print(df.breath.mode()[0])
print(df.temperature.std())
print(df.puls.std())
print(df.otherpuls.std())
print(df.breath.std())
```
