```python
#Задание 1
#Для датафрейма log  создайте столбец source_type по следующим правилам:
#если источник traffic_source равен yandex или google, то в source_type ставится organic
#для источников paid и email из России - ставим ad
#для источников paid и email не из России - ставим other
#все остальные варианты берем из traffic_source без изменений
import pandas as pd
import numpy as np
log = pd.read_csv('visit_log.csv', sep =';')
df = pd.read_csv('visit_log.csv', sep =';')
df.head(100)
```

```python
df.loc[(df.traffic_source == 'yandex') | (df.traffic_source== 'google'), 'source_type'] = 'organic'
#если источник traffic_source равен yandex или google, то в source_type ставится organic
df.head(100)
```

```python
df.loc[(df.traffic_source == 'paid') & (df.region== 'Russia'), 'source_type'] = 'ad'
#для источников paid и email из России - ставим ad
df.head(100)
```

```python
#для источников email из России - ставим ad
df.loc[(df.traffic_source == 'email') & (df.region== 'Russia'), 'source_type'] = 'ad'
df.head(200)
```

```python
#для источников paid из России - ставим ad
df.loc[(df.traffic_source == 'paid') & (df.region== 'Russia'), 'source_type'] = 'ad'
df.head(200)
```

```python
#для источников email не из России - ставим other
df.loc[(df.traffic_source == 'email') & (df.region!= 'Russia'), 'source_type'] = 'other'
df.head(200)
```

```python
#для источников paid не из России - ставим other
df.loc[(df.traffic_source == 'paid') & (df.region!= 'Russia'), 'source_type'] = 'other'
df.head(500)
```

```python
#все остальные варианты берем из traffic_source без изменений
df.loc[(df.traffic_source == 'direct'), 'source_type'] = 'direct'
df.head(8000)
```
